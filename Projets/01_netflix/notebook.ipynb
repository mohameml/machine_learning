```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

```python
# Start coding here! Use as many cells as you like
print(netflix_df.shape)
print(netflix_df.head())
```

```python
# processing data :

df = netflix_df[(netflix_df['release_year'] == 1990) & (netflix_df['type'] == 'Movie')]
df.shape

```

```python
df.head(14)
```

```python
# Visualize the data to view the distributions of durations for movies released in the 1990s

df['duration'].hist()
```

```python
# duration = 0
# for dur in netflix_df['duration'] : 
#     if dur > duration :
#         duration = dur
duration = df['duration'].max()        
print(f"the most frequent movie in the 1990s is for duration {duration}")

films = df.loc[df['duration'] == duration]
print(films)
    
```

```python
short_movie_count = 0 

for dur in df['duration'] :
    if dur  < 90 :
        short_movie_count += 1 

print(f'the number of short actions movies in the 1990s is {short_movie_count}')
```
