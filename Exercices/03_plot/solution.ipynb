```python
# exercice : 
import numpy as np 
import matplotlib.pyplot as plt 

# data : 
data = {f"experience_{i}" : np.random.randn(100) for i in range(4)}

def graphique(dataset):
    """
    pour chaque experience_i : graphe de visulaisation 
    """
    
    nombre_of_courbes = len(dataset)
    x = list(range(0,100))
    # x = np.linspace(-5,5, 100)

    for i in range(nombre_of_courbes):

        plt.subplot(nombre_of_courbes , 1 , i+1)
        plt.plot(x , dataset[f"experience_{i}"])
        plt.title(f"experience_{i}")


    plt.tight_layout()
    plt.show()

        


graphique(data)
```
