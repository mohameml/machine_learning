### Random WalK

```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk_2(n) :

    res = [0]

    for i in range(n) :
        res.append(res[-1] + np.random.choice([-1, 1]))
    return res


def draw_random_walk_2(n) :
    res = random_walk_2(n)
    plt.plot(res)
    plt.hist(np.random.normal(0, 1, n), bins=100, alpha=0.5)
    plt.legend(['Random Walk', 'Normal Distribution'])
    plt.show()

def main_2(n) : 
    draw_random_walk_2(n)

main_2(10000)



```
