```python
# ============================== Exemple 1 ======================

import matplotlib.pyplot as plt 
import numpy as np 
x = np.linspace(0, 10, 100)  # 100 points entre 0 et 10
y = np.sin(x)  # Sinus des points

# Création du graphique
plt.plot(x, y, label='sin(x)')
plt.title('Graphique de sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
# ===========================  Exemple 2 ========================
import matplotlib.pyplot as plt
import numpy as np

# Données
x = np.linspace(0, 10, 100)  # 100 points entre 0 et 10
y1 = np.sin(x)                # Sinus des points
y2 = np.cos(x)                # Cosinus des points

# Création du graphique
plt.plot(x, y1, label='sin(x)', lw=2, ls='-', c='blue')  # Ligne sin(x)
plt.plot(x, y2, label='cos(x)', lw=2, ls='--', c='red')  # Ligne cos(x)

# Personnalisation du graphique
plt.title('Graphiques de sin(x) et cos(x)')
plt.xlabel('x')
plt.ylabel('Valeur')
plt.legend()  # Affiche la légende
plt.grid(True)  # Affiche la grille
plt.show()
```

```python
# ========================== Exemple 3 : subplto ==============================
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(1,100 , 1000)
y = np.exp(x)
z = np.log(x)
plt.figure()
plt.plot(x,y, label="exp" )
plt.plot(x ,z , label="log")
plt.title("fonctions log && exp")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# ===================================== Exemple 4 =============================
import matplotlib.pyplot as plt
import numpy as np

# Création de données d'exemple
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Création d'une figure avec 3 sous-graphes
plt.figure(figsize=(10, 8))

# Premier subplot : Sinus
plt.subplot(3, 1, 1)  # 3 lignes, 1 colonne, 1er subplot
plt.plot(x, y1, label='Sinus', color='blue')
plt.title('Graphique du Sinus')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid()
plt.legend()

# Deuxième subplot : Cosinus
plt.subplot(3, 1, 2)  # 3 lignes, 1 colonne, 2ème subplot
plt.plot(x, y2, label='Cosinus', color='orange')
plt.title('Graphique du Cosinus')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid()
plt.legend()

# Troisième subplot : Tangente
plt.subplot(3, 1, 3)  # 3 lignes, 1 colonne, 3ème subplot
plt.plot(x, y3, label='Tangente', color='red')
plt.title('Graphique de la Tangente')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.ylim(-10, 10)  # Limiter l'axe y pour la tangente
plt.grid()
plt.legend()

# Ajuster l'espacement entre les subplots
plt.tight_layout()

# Afficher la figure
plt.show()

```

```python
# approche POO :
import numpy as np 
import matplotlib.pyplot as plt 
x = np.linspace(-5,5 , 100)
fig , ax = plt.subplots(2 , 1 , sharex=True)
ax[0].plot(x ,np.sin(x)**2 + 1 , label="sin**2 + 1")
ax[0].plot(x , np.cos(x)**2 + 1  , label="cos**2")
ax[0].legend()
ax[1].plot(x , x**2 + np.sqrt(abs(x)) , label="x**2 + sqrt(abs(x))")
ax[1].legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Génération des données
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Création de la figure et des axes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Tracé sur le premier subplot (ax1)
ax1.plot(x, y1, color="blue", label="sin(x)")
ax1.set_title("Courbe de sin(x)")       # Titre pour le premier graphique
ax1.set_xlabel("x")                     # Label de l'axe x
ax1.set_ylabel("sin(x)")                # Label de l'axe y
ax1.legend()                            # Ajout de la légende

# Tracé sur le deuxième subplot (ax2)
ax2.plot(x, y2, color="red", label="cos(x)")
ax2.set_title("Courbe de cos(x)")       # Titre pour le deuxième graphique
ax2.set_xlabel("x")                     # Label de l'axe x
ax2.set_ylabel("cos(x)")                # Label de l'axe y
ax2.legend()                            # Ajout de la légende

# Affichage de la figure
plt.tight_layout()  # Ajuste automatiquement l'espacement entre les subplots
plt.show()
```

### 5. Classification :

```python
# classification : 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris 

df = load_iris()
x = df.data
y = df.target 

print(f"x contient {x.shape[0]} examples  et {x.shape[1]} variables ")
print(f"il y'a  {np.unique(y).size} classes qui sont : {df.target_names}")

plt.scatter(x[:,0] , x[:,1] , c=y , alpha=0.5 , s=x[:,2]*100)
plt.xlabel("longueur sépal")
plt.ylabel("largeur sépal ")

```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='viridis')
plt.xlabel("Longueur sépal")
plt.ylabel("Largeur sépal")
plt.colorbar(ticks=np.unique(y), label="Classe")
plt.show()
```

```python
# Données pour une courbe hélicoïdale
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
x = np.sin(theta)
y = np.cos(theta)

# Création de la figure et de l'axe 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Courbe 3D
ax.plot(x, y, z, label='Courbe hélicoïdale')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
```

```python
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
f = lambda x,y : np.sin(x) + np.cos(x+y)*np.cos(x)
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)
Z = f(X,Y)
ax =plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap="plasma")
```

```python
# Surface pour le fil de fer
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.cos(np.sqrt(x**2 + y**2))

# Création de la figure et de l'axe 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graphe en fil de fer
ax.plot_wireframe(x, y, z, color='black')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
# Données pour l'histogramme
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Histogramme 3D
hist, xedges, yedges = np.histogram2d(x, y, bins=10)
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Dimensions des barres
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition des données
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2  # Fonction à deux variables

ax = plt.axes(projection="3d")
ax.plot_surface(X,Y,Z)

print(np.min(Z))
print(np.max(Z))
```

```python
# Tracé des lignes de niveau
plt.contour(X, Y, Z, levels=5, cmap='viridis')
plt.colorbar()
plt.title('Lignes de niveau de (x^2 + y^2))')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
