```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier

# Gnrer un dataset bidimensionnel (2 classes, 2 features)
X, y = make_classification(n_samples=200, n_features=2, n_classes=2, 
                           n_clusters_per_class=1, n_redundant=0, random_state=42)

# Initialiser et entraner le modle KNN avec k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)

# Dfinir une grille de points pour visualiser la frontire de dcision
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))

# Prdire la classe pour chaque point de la grille
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Tracer la frontire de dcision
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
plt.title("K-Nearest Neighbors (KNN) avec k=5")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
